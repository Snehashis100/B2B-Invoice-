```python
import pandas as pd
```

```python
test_df=pd.read_csv('test_data.csv')
test_df.head()
```

```python
# keeping business_year,posting_date,document_create_date,due_in_date and baseline_create_date as those were the final features 
# given to our model as input
```

```python
test=test_df[['buisness_year','posting_date','document_create_date','due_in_date','baseline_create_date']]
test.head()
```

```python
test.dtypes
```

```python
data_types_dict = {'buisness_year': int}
test = test.astype(data_types_dict)
```

```python
# Converting all the features into proper datetime foramt
```

```python
test['posting_date'] = pd.to_datetime(test['posting_date'])
```

```python
data_types_dict = {'due_in_date': int,'baseline_create_date':'int64'}
test = test.astype(data_types_dict)
```

```python
test['due_in_date'] = pd.to_datetime(test['due_in_date'], format='%Y%m%d')
test['baseline_create_date'] = pd.to_datetime(test['baseline_create_date'], format='%Y%m%d')
test['document_create_date'] = pd.to_datetime(test['document_create_date'], format='%Y%m%d')
test.dtypes
```

```python
# Now converting this datetime into int based on preleptic gregorian ordinal
import datetime as dt
test['posting_date_gregorian']=test['posting_date'].map(dt.datetime.toordinal) 
test['document_create_date_gregorian']=test['document_create_date'].map(dt.datetime.toordinal) 
test['due_in_date_gregorian']=test['due_in_date'].map(dt.datetime.toordinal) 
test['baseline_create_date_gregorian']=test['baseline_create_date'].map(dt.datetime.toordinal) 
```

```python
test.head()
```

```python
test=test.drop(['posting_date','document_create_date','due_in_date','baseline_create_date'],axis=1)
```

```python
test.head()
```

```python
test.buisness_year.value_counts() # the test set has only 2020 as buisness_year
```

```python
# Mapping the buisness_year as 0 if the year is 2019 and 1 if it is 2020 based on the preprocessing done on the train set
test['2020']=test['buisness_year'].map({2020:1,2019:0})
```

```python
test.head()
```

```python
test=test.drop(['buisness_year'],axis=1)
```

```python
# Rearranging the features
test=test[['2020','posting_date_gregorian','document_create_date_gregorian','due_in_date_gregorian','baseline_create_date_gregorian']]
test.head()
```

```python
test.shape
```

```python
#  Feature Scaling on the test set
import pickle
with open('Scaling_obj.pkl','rb') as f:
    scalar=pickle.load(f)
    f.close()
```

```python
test1=scalar.transform(test)
test2=pd.DataFrame(test1,columns=test.columns)
test2.head()
```

```python
# Loading the model pickle
import pickle
with open('Model.pkl','rb') as f:
    model=pickle.load(f)
    f.close()
```

```python
# Prediction on the test set
predictions=model.predict(test2)
```

```python
predictions
```

```python
len(predictions)
```

```python
# As the prediction are on float we need to convert in on nearest int so that we can convert them into date
```

```python
predictions.shape
```

```python
predicted_clear_date=pd.DataFrame(predictions.reshape(-1,1),columns=['predicted_clear_date'])
predicted_clear_date
```

```python
new_df=pd.concat([test,predicted_clear_date],axis=1)
new_df.head()
```

```python
import numpy as np
new_df['predicted_clear_date']=new_df['predicted_clear_date'].apply(lambda x:np.round(x))
```

```python
data_types_dict = {'predicted_clear_date': int}
new_df = new_df.astype(data_types_dict)
```

```python
new_df
```

```python
# Now converting all the gregorian nummbers back to datetime format
def Back_to_date(val):
    ts = pd.Timestamp(year = 2000,  month = 1, day = 1, # ts is a timestamp that it takes as a reference to convert an ordinal into datetime
                  hour = 0, second = 0, tz = 'US/Central')
    return ts.fromordinal(val)
```

```python
final_df=new_df.copy()
```

```python
final_df['posting_date']=final_df['posting_date_gregorian'].apply(Back_to_date)
final_df['document_create_date']=final_df['document_create_date_gregorian'].apply(Back_to_date)
final_df['due_in_date']=final_df['due_in_date_gregorian'].apply(Back_to_date)
final_df['baseline_create_date']=final_df['baseline_create_date_gregorian'].apply(Back_to_date)
final_df['predicted_clear']=final_df['predicted_clear_date'].apply(Back_to_date)
```

```python
# Dropping the gregorian columns
final_df=final_df.drop(['posting_date_gregorian','document_create_date_gregorian','due_in_date_gregorian'
                     ,'baseline_create_date_gregorian','predicted_clear_date'],axis=1)
```

```python
# The 'predicted_clear' represents the clear_date predicted by our model
```

```python
final_df.head()
```

```python
# Predicting the aging bucket based on the clear_date
```

```python
from datetime import date
def get_no_of_days(features):
    due_in_date=features[0]
    predicted_clear=features[1]
    init_year=due_in_date.year
    init_month=due_in_date.month
    init_day=due_in_date.day
    
    final_year=predicted_clear.year
    final_month=predicted_clear.month
    final_day=predicted_clear.day 
    
    date_1 = date(year=final_year, month=final_month, day=final_day)
    date_2 = date(year=init_year, month=init_month, day=init_day)

    date_delta = date_1 - date_2        
    number_of_days = date_delta.days    
    return number_of_days
```

```python
final_df['aging_bucket']=final_df[['due_in_date','predicted_clear']].apply(get_no_of_days,axis=1)
final_df.head()
```

```python
pd.DataFrame(final_df['aging_bucket'].value_counts()).nlargest(15,'aging_bucket')
```

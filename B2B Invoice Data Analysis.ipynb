```python
# Creating a new column that will capture the no of days between the date on which an Invoice is created and the date on which 
# the customer is expected to pay for the  invoice
from datetime import date
def get_no_of_days(features):
    baseline=features[0]
    due_date=features[1]
    init_year=baseline.year
    init_month=baseline.month
    init_day=baseline.day
    
    final_year=due_date.year
    final_month=due_date.month
    final_day=due_date.day 
    
    date_1 = date(year=final_year, month=final_month, day=final_day)
    date_2 = date(year=init_year, month=init_month, day=init_day)

    date_delta = date_1 - date_2        
    number_of_days = date_delta.days    
    return number_of_days
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train2=sc.fit_transform(X_train)
X_test2=sc.transform(X_test)
X_train=pd.DataFrame(X_train2,columns=X_train.columns)
X_test=pd.DataFrame(X_test2,columns=X_test.columns)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score # Importing all the necessary metrics that 
# will be used to find the goodness of the model
```

```python
# This function will return all the metrics values after training itself on the train set and then predicting on the test set
def Model_Performance(x_train,x_test,y_train,y_test,model):
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    r2=r2_score(y_test,pred)
    adj_r2 = 1 -(((1-r2)*(len(x_test)-1))/(len(x_test)-x_test.shape[1]-1))
    mae = mean_absolute_error(y_test, pred)
    mse = mean_squared_error(y_test, pred)
    return r2,adj_r2,mae,mse
```

```python
# Importing various regression models
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
# from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
# from sklearn.svm import SVR
import xgboost as xgb
```

```python
# Linear Regression
linear=LinearRegression()
linear_r2,linear_adj_r2,linear_mae,linear_mse=Model_Performance(X_train,X_test,y_train,y_test,linear)
print(f"r2 score of Linear Regression :{linear_r2}")
print(f"Adjusted r2 score of Linear Regression :{linear_adj_r2}")
print(f"MAE of Linear Regression :{linear_mae}")
print(f"MSE score of Linear Regression :{linear_mse}")
```

```python
# Now converting all the gregorian nummbers back to datetime format
def Back_to_date(val):
    ts = pd.Timestamp(year = 2000,  month = 1, day = 1, # ts is a timestamp that it takes as a reference to convert an ordinal into datetime
                  hour = 0, second = 0, tz = 'US/Central')
    return ts.fromordinal(val)
```

```python
# Predicting the aging bucket based on the clear_date
```

```python
from datetime import date
def get_no_of_days(features):
    due_in_date=features[0]
    predicted_clear=features[1]
    init_year=due_in_date.year
    init_month=due_in_date.month
    init_day=due_in_date.day
    
    final_year=predicted_clear.year
    final_month=predicted_clear.month
    final_day=predicted_clear.day 
    
    date_1 = date(year=final_year, month=final_month, day=final_day)
    date_2 = date(year=init_year, month=init_month, day=init_day)

    date_delta = date_1 - date_2        
    number_of_days = date_delta.days    
    return number_of_days
```

```python
final_df['aging_bucket']=final_df[['due_in_date','predicted_clear']].apply(get_no_of_days,axis=1)
final_df.head()
```

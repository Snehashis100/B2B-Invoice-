```python
import pandas as pd
```

```python
df=pd.read_csv("Invoice Dataset.csv")
df.head()
```

```python
df=df[df['clear_date'].notna()]
df.shape
```

```python
# Finding the differnet categories of the aging bucket for the entire dataset
from datetime import date
def get_no_of_days(features):
    due_in_date=features[0]
    predicted_clear=features[1]
    init_year=due_in_date.year
    init_month=due_in_date.month
    init_day=due_in_date.day
    
    final_year=predicted_clear.year
    final_month=predicted_clear.month
    final_day=predicted_clear.day 
    
    date_1 = date(year=final_year, month=final_month, day=final_day)
    date_2 = date(year=init_year, month=init_month, day=init_day)

    date_delta = date_1 - date_2        
    number_of_days = date_delta.days    
    return number_of_days
```

```python
df['clear_date']=df['clear_date'].apply(lambda x:x.split()[0])
df['clear_date'] = pd.to_datetime(df['clear_date'])
data_types_dict = {'due_in_date': int}
df = df.astype(data_types_dict)
df['due_in_date'] = pd.to_datetime(df['due_in_date'], format='%Y%m%d')
```

```python
df.head()
```

```python
df['aging_bucket']=df[['due_in_date','clear_date']].apply(get_no_of_days,axis=1)
df.head()
```
